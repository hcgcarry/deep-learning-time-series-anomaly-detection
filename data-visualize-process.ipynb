```
# Set normal current interval
alpha = 0.2
temp_list = list()
for df in df_list :
    temp_list = temp_list + list(df['Item001'])
temp_list = [i for i in temp_list if i != 0]
mu = statistics.mean(temp_list)
norm_interval = (mu-alpha, mu+alpha) # justification?
print('normal interval = ', norm_interval)
```

```
# Set break time intervals (df indices)
br_time_1 = (415, 490)
br_time_2 = (700, 730)
```

```
    '''Data Preprocessing'''
# Remove abnormal points
def rm_abn(df) :
    temp_df = pd.Dataframe()
    temp_df = df  
    for i in range(len(temp_df['Item001'])) :
        if norm_interval[0] <= temp_df['Item001'][i] <= norm_interval[1]:
            continue
        elif i in range(br_time_1[0], br_time_1[1]) and temp_df['Item001'][i] == 0 :
            continue
        elif i in range(br_time_2[0], br_time_2[1]) and temp_df['Item001'][i] == 0 :
            continue
        else :
            ran_val = np.random.uniform(norm_interval[0], norm_interval[1])
            temp_df['Item001'][i] = round(ran_val, 2)
    return temp_df

for df in df_list :
  df_list[df_list.index(df)] = rm_abn(df)
```
